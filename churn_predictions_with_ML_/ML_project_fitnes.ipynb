# Описание проекта и данных

**Задача**

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.

Проблема - отток клиентов ( будем считать что клиент больше не с нами, если за последний месяц ни разу не посетил спортзал)

Чтобы бороться с оттоком, отдел по работе с клиентами перевёл в электронный вид множество клиентских анкет. 

Задача — провести анализ и подготовить план действий по удержанию клиентов.


**А именно:**

* научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;

* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;

* проанализировать основные признаки, наиболее сильно влияющие на отток;

* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

1) выделить целевые группы клиентов;

2) предложить меры по снижению оттока;

3) определить другие особенности взаимодействия с клиентами.



**ОПИСАНИЕ ДАННЫХ**

*Данные клиента за предыдущий до проверки факта оттока месяц:*

* 'gender' — пол;

* 'Near_Location' — проживание или работа в районе, где находится фитнес-центр;

* 'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);

* Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);

* 'Phone' — наличие контактного телефона;

* 'Age' — возраст;

* 'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).


Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:


* 'Contract_period' — длительность текущего действующего абонемента (месяц, 6 месяцев, год);

* 'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);

* 'Group_visits' — факт посещения групповых занятий;

* 'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;

* 'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;

* 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

* 'Churn' — факт оттока в текущем месяце.

## Регрессия

* логистическая регрессия

* логистическая регрессия с подбором параметров и кросс-валидацией через GridSearch

```python
# разделим признаки и таргет
random_sate = 0
X = df.drop('churn', axis = 1)
y = df['churn']
# делим на тренировочную и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = random_sate)

# стандартизируем признаки для регрессии
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)

# строим модель и предсказываем
model_reg = LogisticRegression(random_state = random_sate)
model_reg.fit(X_train_st, y_train)
predictions_reg = model_reg.predict(X_test_st)

acc_reg = accuracy_score(y_test, predictions_reg)
prec_reg = precision_score(y_test, predictions_reg)
recall_reg = recall_score(y_test, predictions_reg)

```

```python
import warnings
warnings.filterwarnings("ignore")
#  Теперь найдём наилучшие параметры с помощью кросс - валидации

estimator = LogisticRegression(random_state=random_sate)
param_grid = {'penalty' : ['l1', 'l2', 'elasticnet'],
              'solver' : ['sag', 'saga', 'lbfgs', 'liblinear'],
              'max_iter': [50, 100, 150, 200],
              'class_weight' : [None, 'balanced']
              
             }

model_reg_Grid_cv = GridSearchCV(estimator, param_grid)
model_reg_Grid_cv.fit(X_train_st, y_train)      
```

```python
# посмторим наилучшие параметры для модели
model_reg_Grid_cv.best_params_
```

```python
model_reg.get_params()
```

Можем сравнить параметры обычной регрессии и Grid - LogReg

```python
# выберем наилучшую модель
predictions_reg_Grid_cv = model_reg_Grid_cv.best_estimator_.predict(X_test_st)

acc_reg_cv = accuracy_score(y_test, predictions_reg_Grid_cv)
prec_reg_cv = precision_score(y_test, predictions_reg_Grid_cv)
recall_reg_cv = recall_score(y_test, predictions_reg_Grid_cv)
```

## RandomForest

Будем искать лучшую модель сразу по сетке параметров

```python
params = {'n_estimators' : [50, 100, 150, 200],
         'criterion' : ['gini', 'log_loss', 'entropy'],
          'max_depth' :  list(range(3, 60, 10)),
          'min_samples_split' : list(range(2, 20, 4)),
         }


model_forest = GridSearchCV(estimator=RandomForestClassifier(random_state = 0), param_grid= params)
model_forest.fit(X_train, y_train)
predictions_forest_Grid_cv = model_forest.best_estimator_.predict(X_test)

# посмотрим на параметры
model_forest.best_params_
```

```python
# считаем метрики лдя предсказаний
acc_forest_cv = accuracy_score(y_test, predictions_forest_Grid_cv)
prec_forest_cv = precision_score(y_test, predictions_forest_Grid_cv)
recall_forest_cv = recall_score(y_test, predictions_forest_Grid_cv)
```

## XGB

```python
param_xgb = {
    'booster' : ['gbtree', 'gblinear', 'dart'],
    'learning_rate' : np.arange(0, 0.5, 0.05),
    'max_depth' : [5, 6, 7, 9],
    
    
}


model_xgb = GridSearchCV(estimator = XGBClassifier(random_state = 0, verbosity = 0), param_grid = param_xgb)
model_xgb.fit(X_train, y_train)
predictions_xgb = model_xgb.best_estimator_.predict(X_test)
```

```python
model_xgb.best_params_
```

```python
# считаем метрики лдя предсказаний
acc_xgb = accuracy_score(y_test, predictions_xgb)
prec_xgb = precision_score(y_test, predictions_xgb)
recall_xgb = recall_score(y_test, predictions_xgb)
```

```python
# создадим таблицу с метриками
result_metrics_table = pd.DataFrame.from_dict({'model' : ['LogRegression', 'LogRegression_GridCV',\
                                                          'RandomForestClf_Grid_cv', 'XGBClassifier'],
             'accuracy' : [acc_reg, acc_reg_cv, acc_forest_cv, acc_xgb],
              'precision' : [prec_reg, prec_reg_cv, prec_forest_cv, prec_xgb],
              'recall' : [recall_reg, recall_reg_cv, recall_forest_cv, recall_xgb]             
             })
```

```python
result_metrics_table
```

**Наилучшие результаты показывает XGB**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
sc = StandardScaler()
df_st = sc.fit_transform(df)
linked = linkage(df_st, method = 'ward')
plt.figure(figsize = (16,10))
dendrogram(linked, orientation = 'top')
plt.title('Дендраграмма для df_st')
plt.show()
```

## Смотрим на характеристики классов (+ строим распредениения)

```python
# создадим функцию для попарной визуализации признаков + класс
def plot_xy_claster(data, x, y, title = None):
    
    plt.figure(figsize = (8,5))
    sns.scatterplot(data = data, x = data[x], y = data[y] , hue = 'class', palette="Set1")
    plt.title(title)
    plt.show()
```

Можем посмотреть распределение двух признаков и проерить на принадлежность к классу.

Например, посмотри на соотношение лайфтайма и среднего числа посещений в неделю для тех, кто отвалился и тех, кого мы удержали.

```python
plot_xy_claster(df_claster_predict[df_claster_predict['churn'] == 1], 'lifetime', 'avg_class_frequency_total', \
                title= "Среднее число посещений в неделю / лайфтайм. Кластеризация для ушедших пользователей")
```

```python
plot_xy_claster(df_claster_predict[df_claster_predict['churn'] == 0], 'lifetime','avg_class_frequency_total', \
              title= "Среднее число посещений в неделю / лайфтайм. Кластеризация для оставшихся пользователей")
```

Видим:

* среди оставшихся преобладают 1, 4, 3 классы

* среди оттока преобладают 0 и 2 классы
