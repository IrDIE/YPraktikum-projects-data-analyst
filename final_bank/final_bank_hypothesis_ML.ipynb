**Проверка статистических гипотез**

Здесь и данее выберем уровень статистической значимости альфа = 0,05.

Для каждой гопотезы нам нужно:

* определить статистический тест ( при необходимости сделать проверку на нормальность - Шапиро-тест и qq-plot )

* задать уровень значимости

* провести  тест

* отклонить или оставить нулевую гипотезу




### Гипотеза о баллах кр. скоринга

**H_0 : Среднее значение для балла кредитного скоринга, для оставшихся и отвалившихся пользователей не отличается**

H_1: Среднее балла кредитного скоринга в двух группах значимо отличаются 

Проверка распределения на нормальность ( чтобы понять можем ли применить т-тест )

```python
def get_qq_plots(test_0, test_1):
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    stats.probplot(test_0, dist="norm", plot=pylab)
    plt.title('Для churn = 0')
    
    plt.subplot(1,2,2)
    stats.probplot(test_1, dist="norm", plot=pylab)
    plt.title('Для churn = 1')
    pylab.show()
    

test_0 = df[df['churn'] == 0]['score']
test_1 = df[df['churn'] == 1]['score']
get_qq_plots(test_0, test_1)
```

```python
shapiro_test_0 = stats.shapiro(test_0)
shapiro_test_1 = stats.shapiro(test_1)
print('p_val для test_0 = ', shapiro_test_0.pvalue)
print('p_val для test_1 = ', shapiro_test_1.pvalue)
```

Видим по qq-графику и по p_value что распределения далеки от нормального, не можем применять т-тест.

Будем применять непараметричесткий U-критерий Манна — Уитни

```python
test_0.mean() < test_1.mean()
# будем брать односторонний, alternative='less'
```

```python
stats.mannwhitneyu(test_0, test_1, alternative='less')
```

Не можем отклонить нулевую гипотезу, p_value > 0.05

```python
def get_power_effect(test_0, test_1, alternative):
    effect_size = (test_0.mean() - test_1.mean()) / test_0.mean()
    nobs1 = len(test_0)
    ratio = len(test_1) / len(test_0)
    alpha = 0.05
    power = TTestIndPower().solve_power(effect_size = effect_size, nobs1 = nobs1,ratio = ratio,\
                                    alpha = alpha, alternative = alternative)

    print("Power = ", power)
    print("Effect = ", effect_size)
    
    
get_power_effect(test_0, test_1, alternative = 'smaller')  
```

Видим низкую можность теста, что ожидаемо при таком маленьком эффекте.

### Гипотеза о возрасте

**H_0 : Средний возраст клиентов ушедших и оставшихся не отличается** 

H_1: Средний возраст клиентов ушедших и оставшихся значисо отличается



```python
test_0 = df[df['churn'] == 0]['age']
test_1 = df[df['churn'] == 1]['age']

get_qq_plots(test_0, test_1)
```

```python
shapiro_test_0 = stats.shapiro(test_0)
shapiro_test_1 = stats.shapiro(test_1)
print('p_val для test_0 = ', shapiro_test_0.pvalue)
print('p_val для test_1 = ', shapiro_test_1.pvalue)
```

Видим что для оталившихся пользователей распределение возраста близко к нормальному, а для оставшихся нет.

Попробуем провести и т-тест и непараметрический тест Манна - Уитни.

```python
test_0.mean() < test_1.mean()
# будем брать односторонний, alternative='less'
```

```python
# Найдем p_value в предположении что средний возраст оставшихся меньше
pv_ttest = stats.ttest_ind(test_0, test_1, alternative='less').pvalue
print('p_value with t_test = ', pv_ttest)

pv_mu = stats.mannwhitneyu(test_0, test_1, alternative='less').pvalue
print('p_value with M-U test = ', pv_mu)
```

И тот и другой тест показали P_val << 0.05. Можем отвергнуть нулевую гипотезу.

### Гипотеза о влиянии пола на отток

Нужно сравнить долю успехов ( churn = 0 ) в двух выборках (Мужчины / Женщины ), и понять является ли различие статистически значимым.

H_0: Пол не влияет на отток

H_1: Пол значимо влияет на отток

Будем использовать Z-test



В формуле n₁ и n₂ — размеры двух сравниваемых выборок, то есть количества наблюдений в них; P₁, P₂ — пропорции, наблюдаемые в выборках; P — пропорция в выборке, скомбинированной из двух наблюдаемых; π₁, π₂ — настоящие пропорции в сравниваемых генеральных совокупностях.

https://baguzin.ru/wp/proverka-gipotez-dvuhvyborochnye-kri/




```python
# Закодируем переменные. Ж = 0, М = 1
df['gender_cod'] = preprocessing.LabelEncoder().fit_transform(df['gender'])  
df.head(1)
```

Наши критические Z - значения = +- 1,96



```python
def get_z_value(df, col):
    test_1 = df[df[col] == 0]['churn']
    test_2 = df[df[col] == 1]['churn']
    test = df['churn']

    n1 = len(test_1)
    n2 = len(test_2)
    p1 = (test_1 == 0).sum() / n1
    p2 = (test_2 == 0).sum() / n2
    p = (test == 0).sum() / len(test)
    z = (p1 - p2) / (np.sqrt(p * (1-p) * ( (1 / n1) + ( 1 / n2) )))
    print("Z - value = ", z)
    
    distr = stats.norm(0, 1) 
    p_value = (1 - distr.cdf(abs(z))) * 2
    print('p-значение: ', p_value)
    alpha = 0.05
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )
    return z

z = get_z_value(df, 'gender_cod')
```

Удалось отвергнуть нулевую гипотезу. 



p1 = доля успехов(churn = 0) у Женщин

p2 = доля успехов(churn = 0) у Мужчин

Получили отрицательные z-value, значит p1 < p2. То есть в успехах (= не отвал пользователя)  доля Мужчин значимо больше.
